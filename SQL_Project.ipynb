```python
df2 = pd.read_csv('data-2.0.csv')
df2
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn for pretty colors

# Replacing 'category_name', 'country', and 'totalorderedquantity' with actual column names
grouped_df = df2.groupby(['categoryname', 'country'])['totalorderquantity'].sum().unstack()

# Using Seaborn for pretty colors
colors = sns.color_palette('viridis', n_colors=len(grouped_df.columns))

# Plotting the grouped DataFrame with custom colors
ax = grouped_df.plot(kind='bar', stacked=True, color=colors)

ax.legend(loc='upper right', bbox_to_anchor=(1.20, 1))


# Adding labels and title
plt.xlabel('Category')
plt.ylabel('Total Ordered Quantity')
plt.title('Total Ordered Quantity by Category and Country')


# Show the plot
plt.show()

```

```python
df3 = pd.read_csv('data-3.0.csv')
df3
```

```python
#Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Use seaborn barplot for horizontal bar graph
ax = sns.barplot( x='totalorderquantity',
    y='productname',
    hue='categoryname',
    data=df3,
    palette='viridis',
    ci=None  # Disable error bars
)

# Add value labels to each bar with a check to skip zeros
for p in ax.patches:
    width = p.get_width()
    if width > 0:  # This condition checks for non-zero values
        ax.annotate(f'{width:.0f}',
                    (p.get_width(), p.get_y() + p.get_height() / 2),
                    ha='left',
                    va='center',
                    xytext=(5, 0),
                    textcoords='offset points')

# Add labels and title
plt.xlabel('Quantity ordered')
plt.ylabel('Product Name')
plt.title('Top 3 Products in Top 3 Categories')

# Show the plot
plt.show()
```

```python
df4 = pd.read_csv('data-a.csv')
df4
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Seaborn for pretty colors

# Pivot the DataFrame for plotting
pivot_df = df4.pivot_table(index=['Category', 'productname'], columns='country', values='totalorderquantity')

# Set Seaborn style for better aesthetics
sns.set(style="white")
sns.set_style("ticks")

# Plotting the grouped DataFrame with custom colors
ax = pivot_df.plot(kind='bar', stacked=True, cmap='viridis')

# Adding labels and title
plt.xlabel('Category - Product')
plt.ylabel('Total Ordered Quantity')
plt.title('Top 3 Countries with Highest Orders for Each Product in Each Category')

# Change the legend position and adjust the layout
ax.legend(loc='upper right', bbox_to_anchor=(1.25, 1))

# Show the plot
plt.show()
```
